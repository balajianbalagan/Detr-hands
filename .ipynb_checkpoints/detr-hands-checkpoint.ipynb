```python
class HandPoseDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None):
        """
        Args:
            annotations_file (str): Path to the JSON file with annotations.
            img_dir (str): Directory with all the images.
            transform (callable, optional): Transform to be applied on an image.
        """
        with open(annotations_file, 'r') as f:
            self.annotations = json.load(f)
        self.img_dir = img_dir
        self.image_list = os.listdir(img_dir)
        self.transform = transform

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        keypoint_list = self.annotations[idx]
        img_name = self.image_list[idx]
        img_path = os.path.join(self.img_dir, img_name)
        image = read_image(img_path).float() / 255.0  # Normalize to [0, 1]
        h, w = image.shape[1:]
        keypoints = torch.tensor(keypoint_list, dtype=torch.float32)
        keypoints[:, 0] /= w
        keypoints[:, 1] /= h
        if self.transform:
            image = self.transform(image)
        # Add batch dimension
        image = image.unsqueeze(0)  # (C, H, W) -> (1, C, H, W)
        # Move to the same device as the model
        # Normalize keypoints to [0, 1] based on image dimensions


        return image, keypoints


# Example Usage
annotations_file = "./FreiHAND_pub_v2_eval/evaluation_xyz.json"
img_dir = "./FreiHAND_pub_v2_eval/evaluation/rgb/"

# Optional transforms
transform = Resize((512, 512))

dataset = HandPoseDataset(annotations_file, img_dir, transform)

# Load a sample
image, keypoints = dataset[0]

print(f"Image shape: {image.shape}")
print(f"Keypoints: {keypoints.shape}")
```

```python
import torch
import torch.nn as nn
from torchvision.models import resnet50

class HandPoseDETR(nn.Module):
    def __init__(self, num_keypoints=21, hidden_dim=256, nheads=8,
                 num_encoder_layers=6, num_decoder_layers=6):
        super().__init__()
        
        # ResNet-50 backbone
        self.backbone = resnet50()
        del self.backbone.fc

        # Feature conversion layer
        self.conv = nn.Conv2d(2048, hidden_dim, 1)

        # Transformer
        self.transformer = nn.Transformer(
            hidden_dim, nheads, num_encoder_layers, num_decoder_layers
        )

        # Prediction head (adjusted output shape)
        self.hand_pose_head = nn.Linear(hidden_dim, 3)  # Output 3 values per keypoint (x, y, z)

        # Object query (one for each keypoint)
        self.query_pos = nn.Parameter(torch.rand(num_keypoints, hidden_dim))

        # Positional encodings for spatial information
        self.row_embed = nn.Parameter(torch.rand(50, hidden_dim // 2))
        self.col_embed = nn.Parameter(torch.rand(50, hidden_dim // 2))

    def forward(self, images):
        # Extract features from ResNet-50
        x = self.backbone.conv1(images)
        x = self.backbone.bn1(x)
        x = self.backbone.relu(x)
        x = self.backbone.maxpool(x)

        x = self.backbone.layer1(x)
        x = self.backbone.layer2(x)
        x = self.backbone.layer3(x)
        x = self.backbone.layer4(x)

        # Convert to 256 feature planes
        h = self.conv(x)

        # Compute positional encodings
        H, W = h.shape[-2:]
        pos = torch.cat([
            self.col_embed[:W].unsqueeze(0).repeat(H, 1, 1),
            self.row_embed[:H].unsqueeze(1).repeat(1, W, 1),
        ], dim=-1).flatten(0, 1).unsqueeze(1)

        # Transformer forward pass
        h = self.transformer(
            pos + 0.1 * h.flatten(2).permute(2, 0, 1),  # Encoder input
            self.query_pos.unsqueeze(1)  # Decoder input (queries)
        ).transpose(0, 1)  # Shape: (batch_size, num_queries=21, hidden_dim)

        # Predict (x, y, z) for each keypoint
        keypoints = self.hand_pose_head(h)  # Shape: (batch_size, 21, 3)

        return keypoints
```

## PASS DATA LOADER TO MODEL

```python

class DETRKeypointModel(nn.Module):
    def __init__(self, num_keypoints=21, hidden_dim=256):
        super().__init__()

        # Load Pretrained DETR with ResNet50 backbone
        config = DetrConfig.from_pretrained("facebook/detr-resnet-50")
        self.detr = DetrForObjectDetection(config)

        # Modify the output layers to predict keypoints instead of bounding boxes
        self.keypoint_head = nn.Linear(hidden_dim, num_keypoints * 3)  # (x, y, z)

    def forward(self, images):
        outputs = self.detr.model(images)  # Pass image through DETR
        hs = outputs.last_hidden_state  # Transformer output
        keypoints = self.keypoint_head(hs[:, 0, :])  # Extract keypoints from output
        return keypoints.view(-1, 21, 3)  # Reshape to (Batch, 21, 3)

# Create the model
model = DETRKeypointModel()
model = model.cuda() if torch.cuda.is_available() else model
print(model)
```

```python
class HandPoseDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None):
        with open(annotations_file, 'r') as f:
            self.annotations = json.load(f)
        self.img_dir = img_dir
        self.image_list = os.listdir(img_dir)
        self.transform = transform or Resize((512, 512))

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        keypoint_list = self.annotations[idx]
        img_name = self.image_list[idx]
        img_path = os.path.join(self.img_dir, img_name)

        # Load image
        image = read_image(img_path).float() / 255.0  # Normalize
        image = self.transform(image)  # Resize if needed

        # Fix here: Get correct height and width
        _, h, w = image.shape  # (C, H, W)

        # Normalize keypoints (assuming keypoints are in pixel values)
        keypoints = torch.tensor(keypoint_list, dtype=torch.float32)
        keypoints[:, 0] /= w  # Normalize x
        keypoints[:, 1] /= h  # Normalize y
        keypoints[:, 2] /= 1  # Assume z is already normalized

        return image, keypoints

# Load dataset
dataset = HandPoseDataset("./FreiHAND_pub_v2_eval/evaluation_xyz.json", "./FreiHAND_pub_v2_eval/evaluation/rgb/")
```

## MODEL TRAINING

```python
# Create a logs directory
os.makedirs("logs", exist_ok=True)

# Set up logging
logging.basicConfig(
    filename="logs/training.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="w",
)

# Hyperparameters
batch_size = 10
lr = 1e-4
num_epochs = 1000

# Dataloader
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Loss function
loss_fn = nn.MSELoss()  # Since keypoints are continuous values
optimizer = optim.AdamW(model.parameters(), lr=lr)

# Training loop
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

logging.info(f"Starting training on {device}")
logging.info(f"Batch size: {batch_size}, Learning rate: {lr}, Epochs: {num_epochs}")

def compute_accuracy(predictions, targets):
    """Calculate average Euclidean distance between predicted and true keypoints."""
    with torch.no_grad():
        distances = torch.norm(predictions - targets, dim=-1)  # Compute Euclidean distance per keypoint
        avg_distance = distances.mean().item()  # Average across all keypoints
    return avg_distance

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    total_accuracy = 0
    epoch_losses = []
    epoch_accuracies = []

    for batch_idx, (images, keypoints) in enumerate(dataloader):
        images, keypoints = images.to(device), keypoints.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        
        loss = loss_fn(outputs, keypoints)  # Compute loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        epoch_losses.append(loss.item())

        # Compute accuracy
        accuracy = compute_accuracy(outputs, keypoints)
        total_accuracy += accuracy
        epoch_accuracies.append(accuracy)

        # Log every N batches
        if batch_idx % 10 == 0:
            avg_loss = sum(epoch_losses[-10:]) / len(epoch_losses[-10:])  # Last 10 losses
            avg_accuracy = sum(epoch_accuracies[-10:]) / len(epoch_accuracies[-10:])  # Last 10 accuracy
            logging.info(
                f"Epoch [{epoch+1}/{num_epochs}] | Batch [{batch_idx}/{len(dataloader)}] | "
                f"Loss: {loss.item():.6f} | Avg Loss (last 10): {avg_loss:.6f} | "
                f"Accuracy: {accuracy:.6f} | Avg Accuracy (last 10): {avg_accuracy:.6f}"
            )

    avg_epoch_loss = total_loss / len(dataloader)
    avg_epoch_accuracy = total_accuracy / len(dataloader)

    logging.info(f"Epoch [{epoch+1}/{num_epochs}] completed | Avg Loss: {avg_epoch_loss:.6f} | Avg Accuracy: {avg_epoch_accuracy:.6f}")
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_epoch_loss:.4f}, Accuracy: {avg_epoch_accuracy:.4f}")

print("Training complete!")
logging.info("Training complete!")
```

## EVALUATION

```python
import torch
import numpy as np

def evaluate_model(model, dataloader, device):
    model.eval()  # Set model to evaluation mode

    mse_loss = 0
    mae_loss = 0
    l2_distances = []  # Store L2 errors per keypoint

    with torch.no_grad():
        for images, keypoints in dataloader:
            images, keypoints = images.to(device), keypoints.to(device)

            # Forward pass
            predictions = model(images)  # Shape: (batch, 21, 3)

            # Compute errors
            mse_loss += torch.nn.functional.mse_loss(predictions, keypoints, reduction="sum").item()
            mae_loss += torch.nn.functional.l1_loss(predictions, keypoints, reduction="sum").item()

            # Compute L2 distances per keypoint
            batch_l2_distances = torch.norm(predictions - keypoints, dim=-1)  # Euclidean distance
            l2_distances.extend(batch_l2_distances.cpu().numpy())  # Store per-keypoint distances
            break
    # Compute final metrics
    num_samples = len(dataloader.dataset)  # Total images
    avg_mse = mse_loss / num_samples
    avg_mae = mae_loss / num_samples
    avg_l2_distance = np.mean(l2_distances)  # Mean L2 error per keypoint

    print(f"Evaluation Results:")
    print(f"🔹 Mean Squared Error (MSE): {avg_mse:.6f}")
    print(f"🔹 Mean Absolute Error (MAE): {avg_mae:.6f}")
    print(f"🔹 Average L2 Distance per Keypoint: {avg_l2_distance:.6f}")

    return avg_mse, avg_mae, avg_l2_distance
```

```python
# Create a test dataloader (use a separate validation set if available)
test_dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

# Run evaluation
evaluate_model(model, test_dataloader, device)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Example Camera Intrinsic Matrix (Modify this for your dataset)
K = np.array([
    [600, 0, 320],  # fx, 0, cx
    [0, 600, 240],  # 0, fy, cy
    [0, 0, 1]       # 0,  0,  1
])

def project_keypoints_3d_to_2d(keypoints_3d, K):
    """
    Project 3D world coordinates (X, Y, Z) into 2D image coordinates (u, v).
    Uses matrix multiplication for efficiency.
    
    Args:
        keypoints_3d: (N, 3) numpy array of keypoints in world space.
        K: (3, 3) numpy array (camera intrinsics).
    
    Returns:
        keypoints_2d: (N, 2) numpy array of projected 2D coordinates.
    """
    keypoints_3d = keypoints_3d.T  # Transpose to shape (3, N)
    keypoints_2d_hom = K @ keypoints_3d  # Multiply with camera intrinsics
    keypoints_2d = keypoints_2d_hom[:2] / keypoints_2d_hom[2]  # Divide by Z to normalize
    return keypoints_2d.T  # Shape: (N, 2)

def visualize_keypoints(image, keypoints_2d):
    """
    Display image with projected keypoints.
    """
    plt.imshow(image.permute(1, 2, 0).cpu().numpy())  # Convert (C, H, W) to (H, W, C)
    plt.scatter(keypoints_2d[:, 0], keypoints_2d[:, 1], c='red', marker='o', label="Projected Keypoints")
    plt.legend()
    plt.show()

# Load an image and keypoints (Example)
dataset = HandPoseDataset(annotations_file, img_dir)
image, keypoints_3d = dataset[0]  # Get sample image and keypoints

# Convert to numpy
keypoints_3d = keypoints_3d.cpu().numpy()  # Shape: (21, 3)

# Project 3D keypoints to 2D
keypoints_2d = project_keypoints_3d_to_2d(keypoints_3d, K)

# Visualize
visualize_keypoints(image, keypoints_2d)
```
